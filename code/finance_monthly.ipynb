```python
def set_excel_color (df, path):
    import pandas as pd
    from openpyxl import load_workbook
    # Create a Pandas Excel writer using XlsxWriter as the engine.
    writer = pd.ExcelWriter(path, engine='xlsxwriter')

    # Convert the dataframe to an XlsxWriter Excel object.
    df.to_excel(writer, sheet_name='Sheet1', index=False, encoding='utf8')
    # Get the xlsxwriter workbook and worksheet objects.
    workbook  = writer.book
    worksheet = writer.sheets['Sheet1']

    # Add a format. Light red fill with dark red text.
    format_bg_orange = workbook.add_format({'bg_color': '#FFA500'}) # 橘 ffa500,FF9933 
    format_bg_yellow = workbook.add_format({'bg_color': '#FFFFBB'}) # 黃
    format_bg_wheat = workbook.add_format({'bg_color': '#fadbb9'}) # 皮膚色、小麥色
    format_bg_red = workbook.add_format({'bg_color': '#FFC7CE'}) # 紅
    
    start_row = 0
    end_row = len(df)
    
    if 'finance_yearly' in path:
        worksheet.conditional_format(start_row, 0, end_row, 0,  # 2021 預估本益比、預估max min 股價 
                                     {'type':'cell',
                                    'criteria': '=',
                                    'value':'"2021"',
                                    'format':format_bg_wheat})
    elif 'finance_quarterly_' in path:
        worksheet.conditional_format(start_row, 3, end_row, 3,  # 營收/季 > 40億
                                     {'type':'cell',
                                    'criteria': '>',
                                    'value':40,
                                    'format':format_bg_wheat})
        worksheet.conditional_format(start_row, 14, end_row, 14,  # 母公司收入 < 60%
                                     {'type':'cell',
                                    'criteria': '<',
                                    'value':60,
                                    'format':format_bg_red})
        worksheet.conditional_format(start_row, 17, end_row, 17,  # 毛利率 > 15%
                                     {'type':'cell',
                                    'criteria': '>',
                                    'value':15,
                                    'format':format_bg_wheat})
        worksheet.conditional_format(start_row, 18, end_row, 18,  # 營益率 > 10%
                                     {'type':'cell',
                                    'criteria': '>',
                                    'value':10,
                                    'format':format_bg_wheat})
        worksheet.conditional_format(start_row, 24, end_row, 24,  # 股利占EPS比例(%) > 60%
                                     {'type':'cell',
                                    'criteria': '>',
                                    'value':60,
                                    'format':format_bg_wheat})
        worksheet.conditional_format(start_row, 25, end_row, 25,  # 殖利率 > 6%
                                     {'type':'cell',
                                    'criteria': '>',
                                    'value':6,
                                    'format':format_bg_wheat})
        worksheet.conditional_format(start_row, 26, end_row, 26,  # 去年同季營收成長 >20%
                                     {'type':'cell',
                                    'criteria': '>',
                                    'value':20,
                                    'format':format_bg_wheat})
    else:
        # =================== 漲跌% ============================================
        worksheet.conditional_format(start_row, 8, end_row, 9,  # 成長>20%
                                     {'type':'cell',
                                    'criteria': '>',
                                    'value':20,
                                    'format':format_bg_wheat})
        worksheet.conditional_format(start_row, 12, end_row, 12,  # 成長>20%
                                     {'type':'cell',
                                    'criteria': '>',
                                    'value':20,
                                    'format':format_bg_wheat})
    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
```

```python
def get_yearly(stock_id, lst_EPS, this_year):
#     stock_id = '2330' # 單位(元)
    url = 'https://www.twse.com.tw/exchangeReport/FMNPTK?response=html&stockNo={}'.format(stock_id)
    df_yearly = pd.read_html(url)[0]
    df_yearly = df_yearly.iloc[:, [0, 4, 6, 8]]
    df_yearly.columns = ['year', 'highest', 'lowest', 'average']
    df_yearly['year'] = df_yearly['year'].apply(lambda x: x+ 1911)
    df_yearly = df_yearly[df_yearly['year']>2014]
    
    df_yearly['year'] = df_yearly['year'].astype(str)
    df_yearly = df_yearly.sort_values('year', ascending=False)
    df_yearly.set_index('year', inplace=True)
    
    for item in lst_EPS:
        df_yearly.loc[item['year'], 'EPS'] = item['eps']
        df_yearly.loc[item['year'], '本益比_max'] = round(df_yearly.loc[item['year'], 'highest'] / item['eps'], 1)
        df_yearly.loc[item['year'], '本益比_min'] = round(df_yearly.loc[item['year'], 'lowest'] / item['eps'], 1)
    
    pe_max = math.floor(max(df_yearly['本益比_max'][:4]))
    pe_min = math.ceil(min(df_yearly['本益比_min'][:4]))
    df_yearly.loc[this_year, 'highest'] = df_yearly.loc[this_year, 'EPS'] * pe_max
    df_yearly.loc[this_year, 'lowest'] = df_yearly.loc[this_year, 'EPS'] * pe_min
    df_yearly.loc[this_year, '本益比_max'] = pe_max
    df_yearly.loc[this_year, '本益比_min'] = pe_min
    
    df_yearly.reset_index(inplace=True)
    df_yearly.sort_values('year', ascending=False, inplace=True)
    df_yearly.insert(0, 'name', value=df_today.loc[stock_id, 'name'])
    df_yearly.insert(0, 'stock_id', value=stock_id)
    
    global df_yearly_all
    df_yearly_all = df_yearly_all.append(df_yearly, ignore_index=True)
#     df_yearly.to_excel(path_yearly + 'yearly_{}.xlsx'.format(stock_id), index=False)
```

# 計算歷年同季的營收成長率

```python
def get_growth(this_quarter):
    this_year = int(this_quarter[:4])
    
    # 估算年度EPS
    lst_quarter = [str(this_year) + 'Q' + str(i) for i in range(1,5)]
    num_quarter = len(df[df.index.isin(lst_quarter)]) # 已公布的季報數量
    EPS_year = round(sum(df[df.index.isin(lst_quarter)]['每股盈餘(元)']) / num_quarter * 4, 2)
    
    if not stock_id in df_today.index:
        df_OTC = get_OTC(stock_id)
        df.loc[this_quarter, '股價'] = df_OTC.iloc[0]['Close']
    else:
        df.loc[this_quarter, '股價'] = df_today.loc[stock_id, 'close']
    df.loc[this_quarter, 'EPS(Y)'] = EPS_year
    df.loc[this_quarter, '本益比(Y)'] = round(df.loc[this_quarter, '股價'] / EPS_year,2) # 本益比 = 股價 / 年度EPS
    if (this_year > 2014) & (list(df_dividend.index).count(stock_id)>1) :
#         print(this_year, stock_id, len(df_dividend[df_dividend['year']==this_year]))
        df.loc[this_quarter, '股利_avg'] = df_dividend[df_dividend['year']==this_year].loc[stock_id, 'cash']
    else:
        df.loc[this_quarter, '股利_avg'] = 0

    lst_year = set(df.index.to_series().apply(lambda x: int(x[:4])))
    min_year = min(lst_year)   
    # 算過去5年同季的成長率
    for i in range(1,6):
        old_year = this_year - i
        if old_year < min_year+1:
            break
        old_quarter = this_quarter.replace(str(this_year),str(old_year))

        new_income = df.loc[this_quarter, '營業收入']
        old_income = df.loc[old_quarter, '營業收入']
        growth = round((new_income / old_income - 1) * 100,1)
        
        df.loc[this_quarter, str(old_year)] = growth
#         print(old_quarter, new_income, old_income, growth)
```

```python
def get_last_eps(this_quarter):
    this_year = int(this_quarter[:4])
    # 今年股利 是根據去年EPS發的
    eps_lastyear = 0
    if (str(this_year-1)+'Q4') in df.index:
        eps_lastyear = df.loc[str(this_year-1)+'Q4', 'EPS(Y)']
        
    df.loc[this_quarter, '股利(%)'] = round(df.loc[this_quarter, '股利_avg'] / eps_lastyear*100, 1)
    df.loc[this_quarter, '殖利率(%)'] = round(df.loc[this_quarter, '股利_avg'] / df.loc[this_quarter, '股價']*100, 1)
        
```

```python
folder_name = 'trust/'
root = 'D:/Mia/Stock/'
path_quarterly = root + 'quarterly/'
path_monthly = root + 'monthly/'
path_check = path_quarterly + folder_name+'finance_quarterly_{}.xlsx'.format(s_today)
path_company_stock_num = root + 'company_list.xlsx'

today = datetime.today()
cur_year = today.year
cur_month = today.month
dic_quarter = {1:[1, 2, 3], 2:[4, 5, 6], 3:[7, 8, 9], 4:[10, 11, 12]}

def get_quarter(month):
    for k, v in dic_quarter.items():
        if cur_month-1 in v:
            return k
cur_quarter = get_quarter(cur_month)
print(cur_quarter)

def get_quarter_income(num_year, quarter): # 單位：億
    lst_income = []
    for mon in dic_quarter[quarter]:
        file_monthly = path_month_finance + 't21sc03_{year}_{month}.csv'.format(year=num_year-1911, month=mon)
        if os.path.exists(file_monthly):
            df_month = pd.read_csv(file_monthly, dtype={'公司代號':str})
            df_month = df_month[df_month['公司代號']==stock_id]
            # 單位：千>>億
            lst_income.append(round(df_month.iloc[0,5] / 100000, 2))
    income_sum = sum(lst_income) / len(lst_income) * 3
    return income_sum

def get_quarter_finance(stock_id, df, year, quarter):
    # 估算新一季的EPS
    df_new = pd.DataFrame([{'stock_id':stock_id, 'name':df.iloc[0,1], 'quarter':str(year)+'Q'+str(quarter)}])

    num_100million = 100000000 # 億
    if not stock_id in df_today.index: # 上櫃公司
        return df_new
    
    df_new['普通股數'] = df_company_stock_num.loc[stock_id, '發行股數(千股)']

    # 手動填寫
    # 用月報的營收 估算季報的營收
    df_new['營業收入'] = get_quarter_income(year, quarter)

    df_new['營業費用'] = round(np.mean(df.loc[df.index[:3],'營業費用']),2) # 近三季的平均 營業費用
    df_new['業外收入'] = round(np.mean(df.loc[df.index[:3],'業外收入']),2) # 近三季的平均 業外收入
    df_new['稅率'] = round(np.mean(df.loc[df.index[:3],'稅率']),2) # 近三季的平均 稅率
    df_new['毛利率(%)'] = round(np.mean(df.loc[df.index[:3],'毛利率(%)']),2) # 近三季的平均 毛利率
    df_new['股利(%)'] = round(np.mean(df.loc[df.index[4],'股利(%)']),2) # 
    
    # 計算
    df_new['營業毛利'] = round(df_new.loc[0, '營業收入'] * df_new.loc[0, '毛利率(%)']/100,2) 
    df_new['營業利益'] = round(df_new.loc[0, '營業毛利'] - df_new.loc[0, '營業費用'],2) 
    df_new['稅前淨利'] = round(df_new.loc[0, '營業利益'] + df_new.loc[0, '業外收入'],2) 
    df_new['稅後淨利'] = round(df_new.loc[0, '稅前淨利'] * (1 - df_new.loc[0, '稅率']),2) 
    df_new['每股盈餘(元)'] = round(df_new.loc[0, '稅後淨利'] * num_100million / df_new.loc[0, '普通股數'],2)
    eps_prep = 0
    if quarter > 1:
        eps_prep = sum(df.loc[df.index[:quarter-1],'每股盈餘(元)'])
    df_new['EPS(Y)'] = round((df_new.loc[0, '每股盈餘(元)']+eps_prep)/quarter*4 ,2) # 估算整年EPS
    df_new['股利_avg'] = round(df_new.loc[0, 'EPS(Y)'] * df_new.loc[0, '股利(%)']/100, 2) 
    df_new['股價'] = df_today.loc[stock_id, 'close']
    df_new['本益比(Y)'] = round(df_new.loc[0, '股價'] / df_new.loc[0, 'EPS(Y)'], 0) 
    df_new['殖利率(%)'] = round(df_new.loc[0, '股利_avg'] / df_new.loc[0, '股價']*100, 2) 
    return df_new
```

```python
df_all = df_quarterly_all 
lst_stock_id = list(df_all['stock_id'].unique())
print('有', len(lst_stock_id), '筆')

# 取得發行股數
df_company_stock_num = pd.read_excel(path_company_stock_num, engine='openpyxl', dtype={'stock_id':str})
df_company_stock_num.set_index('stock_id', inplace=True)


# 預測EPS
df_estimate = pd.DataFrame(columns = list(df_all.columns)+['普通股數'])

for stock_id in lst_stock_id:
    df = df_all[df_all['stock_id']==stock_id]
    df_new_quarter = get_quarter_finance(stock_id, df, cur_year, cur_quarter)
    
    df_concat = pd.concat([df, df_new_quarter])
    df_concat.sort_values('quarter', ascending=False, inplace=True)
    df_estimate = df_estimate.append(df_concat)
    
df_estimate.to_excel(path_check.replace('finance_quarterly', 'estimate'), index=False)
df_estimate.head(2)
```
